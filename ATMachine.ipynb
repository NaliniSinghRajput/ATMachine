```python
x = int(input("How much cash do you want to keep in your account?"))

print("okay so your current balance is ",x)
print("1. Check the Balance\n2.Deposit Money\n3.Withdraw Money\n")

y = int(input("Now what do you wish to do?"))

if y == 1:
    print("Your current balance is ",x)
elif y == 2:
    d = int(input("How much do you want to deposit?"))
    deposit = x + d
    x = deposit 
    print("Your new balance is: ",x)
elif y == 3:
    w = int(input("How much would you like to withdraw?"))
    if w > x:
        print("Insufficient Funds!")
    elif w == x:
        print("Your current balance is 0")
    else:
        balance = x - w
        x = balance
        print("This makes your current balance: ", x)

    
        
    
```
